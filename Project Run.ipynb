```python
import scanpy as sc
from cnatool.infer import infer_cna
import numpy as np


# Load your query dataset
adata_query = sc.read_h5ad("PBMC_simulated_cnas_041025.h5ad")
if not isinstance(adata_query.X, np.ndarray):
    adata_query.X = adata_query.X.toarray()
adata_query.X = adata_query.X.astype(np.float32)


disclusters, cna_summary, cna_long = infer_cna(
    adata_query,
    adata_ref=None,
    ref_clustered_file=None,
    cluster_label="cluster",
    overlap_threshold=0.5,
    min_cells=20,
    dispersion_ratio_threshold=2.0,
    min_genes=15,
    z_threshold=1.8,
    region_col="region",             # make region_col explicit
    max_centered_threshold=3,
    window_size=101,
    verbose=True
)

print("Flagged clusters:", disclusters)
print("CNA summary:")
print(cna_summary.head())
print("Per-cell CNA long table:")
print(cna_long.head())
```

```python
cna_long.to_csv("o5m15dip2m10z2.csv", index=False)
```

```python
import scanpy as sc
adata_ground_truth = sc.read_h5ad("PBMC_simulated_cnas_041025.h5ad")

from cnatool.cna_report import cna_region_confusion_overlap
summary_df = cna_region_confusion_overlap(adata_ground_truth, cna_long)
print(summary_df['match'].value_counts())
print(summary_df.head())
```

```python
summary_df.to_csv("o5m15dip2m10z2_confusion.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_heatmap(cm_df, title="CNA Confusion Matrix"):
    plt.figure(figsize=(5, 4))
    sns.heatmap(summary_df, annot=True, fmt='d', cmap='Blues', cbar=True)
    plt.xlabel("Predicted CNA Call")
    plt.ylabel("True CNA Call")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# Call the function:
plot_confusion_heatmap(summary_df)
```
